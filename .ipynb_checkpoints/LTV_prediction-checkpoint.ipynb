## Introduce the Data

### What are the features?
- **Unnamed: 0:** it is index column
- **userID_anonymized:** each user has unique userid
- **arrival_time:** the time when a user opens the game
- **event_time:** the time when an event occurs
- **eventType:** there exists "level_event", "interstitial_impression", "rewarded_impression" and "banner_impression"
                 meaning that users' stage during their journey
- **event_data:** data gathered during each eventType
                duration: how long is the level event in milliseconds
                level: unique level id
                progres: users progress on the game, if progress is 10 then user is on the 10. level however 
                         "level" can be different as level id is an internal identification of levels
                play_count: basically same as progress, since for this specific game there is no lose state
                status: simple indicator of the in-level progress
                        0: user lose the level
                        1: user win the level
                        2: user start level, duration= 0
                revenue: revenue generated by impression
- **platform:** mobile phone or tablet system
- **session_no:** unique session number of user
- **wifi_state:** connection source
- **model:** the model of the device used to enter the game
- **operating_system:** version of the mobile phone or tablet system
- **totalrevenue:** total revenue made by each user after 30 days since the arrival_time

```python
# event_data contains important subdata. let's make them into columns so that we can analyze easily

def convert_to_dict(s):
    dic = dict()
    if s == "{}":
        pass
        
    try:
        modified_string = s.replace('{', '').replace('}', '')
        string_list = modified_string.split(",")
       
        for i in string_list:
            x = re.split("=|:", i)
            dic[x[0].strip()] = x[1]
    except:
        print(s)

    return dic


# apply the above function / converted the rows in the event_data to the dictionary
df["event_data"] = df["event_data"].apply(convert_to_dict)

# convert dictionary to column
def pop_values_tolist(df):
    return df.join(pd.DataFrame(df.pop("event_data").values.tolist()))

# assigning new columns to dataframe
df = pop_values_tolist(df)
```

## Outlier Detection

```python
# functions for finding and removing outliers

def outlier_thresholds(dataframe, variable):
    quartile1 = dataframe[variable].quantile(0.01)
    quartile3 = dataframe[variable].quantile(0.99)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    return low_limit, up_limit

def replace_with_thresholds(dataframe, variable):
    low_limit, up_limit = outlier_thresholds(dataframe, variable)
    dataframe.loc[(dataframe[variable] < low_limit), variable] = low_limit
    dataframe.loc[(dataframe[variable] > up_limit), variable] = up_limit
```

```python
print("session_no          low limit:",outlier_thresholds(df,"session_no")[0],"              ","up_limit:",outlier_thresholds(df,"session_no")[1])
print("wifi_state          low limit:",outlier_thresholds(df,"wifi_state")[0],"               ","up_limit:",outlier_thresholds(df,"wifi_state")[1])
print("totalrevenue        low limit:",outlier_thresholds(df,"totalrevenue")[0]," ","up_limit:",outlier_thresholds(df,"totalrevenue")[1])
print("duration            low limit:",outlier_thresholds(df,"duration")[0],"      ","up_limit:",outlier_thresholds(df,"duration")[1])
print("level               low limit:",outlier_thresholds(df,"level")[0],"            ","up_limit:",outlier_thresholds(df,"level")[1])
print("progress            low limit:",outlier_thresholds(df,"progress")[0],"             ","up_limit:",outlier_thresholds(df,"progress")[1])
print("status              low limit:",outlier_thresholds(df,"status")[0],"               ","up_limit:",outlier_thresholds(df,"status")[1])
print("revenue             low limit:",outlier_thresholds(df,"revenue")[0],"up_limit:",outlier_thresholds(df,"revenue")[1])
print("arrival_time_hours  low limit:",outlier_thresholds(df,"arrival_time_hours")[0],"              ","up_limit:",outlier_thresholds(df,"arrival_time_hours")[1])
print("event_time_hours    low limit:",outlier_thresholds(df,"event_time_hours")[0],"              ","up_limit:",outlier_thresholds(df,"event_time_hours")[1])
```

```python
outlier_removal_list = ["session_no","wifi_state","totalrevenue","duration","level","progress",
                         "status","revenue","arrival_time_hours","event_time_hours"]

for outlier in outlier_removal_list:
    replace_with_thresholds(df,outlier)
```

## Handling missing data

- Models can not handle missing data


- Simplest solution
    - Remove observations/features that have missing data
    

- But, removing missing data can introduce a lot of issues
    - Data is randomly missing: potentially lose a lot of your data
    - Data is non-randomly missing: in addition to losing data, you are also introducing potential biases
    - Usually, this is a poor solution


- An alternative solution is to use imputation
    - Replace missing value with another value
    - Strategies: mean, median, highest frequency value of given feature
    - There are also some different methods such as SimpleImputer, Itaretive Imputer, KNNImputer etc.
        Normally I wanted to use the KNNImputer method. But due to the problem caused by the processor of my computer, 
        I will fill it with the median. Because the computer turns itself off and on due to overheating.

```python
cols = ["duration", "level", "progress", "status", "revenue"]
for col in cols:
    df[cols] = df[cols].fillna(df[cols].median())  
```

### PCA
- Principal component analysis (PCA) is a technique that transforms a dataset of many features into principal components that "summarize" the variance that underlies the data


- Each principal component is calculated by finding the linear combination of features that maximizes variance, while also ensuring zero correlation with the previously calculated principal components 


- Use cases for modeling:
    - One of the most common dimensionality reduction techniques
    - Use if there are too many features or if observation/feature ratio is poor
    - **Also, potentially good option if there are a lot of highly correlated variables in your dataset**


- Unfortunately, PCA makes models a lot harder to interpret
